````python
#Punto 2.4
arr = np.random.randint(0,99, size = (11,11))
arr

#Función para realizar la normalización del array
def normalize(arr):
    """Función para realizar normalización de valores dentro de un array
    Recibe como parametros un array y calcula la normalización de sus valores por medio de ecuaciones matematicas"""
    norm_arr=[] #Creo una lista para alamcenar los valores normalizados
    media = np.mean(arr) #Calculo media
    desvStan = np.std (arr)  #Calculo desviación estandar
    for i in arr: #Recorro los valores del array
        norM = (i - media) / desvStan #Aplico la normalización
        norm_arr.append(norM) #Agrego los valores normalizados a norm_arr
    return norm_arr

#normalize(arr)

#Punto 2.5
arr = np.random.randint(0,99, size = (11,11)) #Array de 11 x 11 con valores entre 1 - 100
arr

unicos = np.unique(arr, return_index=False, return_inverse=False, return_counts=True, axis=None) #Determino la frecuencia del valor que mas se repite
#unicos = np.unique(arr)
unicos
````

````python
file = open("plethysmography.txt", "r")

plesti = file.read().split()

plesti = np.array([float(valor) for valor in plesti])
plesti

def mediaMovil(matriz, numElementos:int):
    """Recibe matriz y numero de elementos para calcular la media"""
    return np.convolve(matriz, np.ones(numElementos), 'valid') / numElementos


mediaMovil(plesti,100)
#plesti
````

¿Qué tipo de filtro es una media movil?
El filtro de media móvil es un filtro del tipo pasa bajo, ya que deja pasar los valores
por debajo de la media calculada, siendo una solución rápida y fácil para 
filtrar ruido de alta frecuencia
